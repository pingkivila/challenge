# **VISUALISASI**

**MERGE CHALLENGE A**

**MERGE CHALLENGE B**

```python
# gabungkan tabel challenge B start dengan challenge B complate
Chall_B = Chall_B_start.merge(Chall_B_comp, how='outer', on=['Name','Username'])
Chall_B
```

```python
pie_data = Chall_A['Total_Action_x'].value_counts(normalize=True).values * 100
pie_label = Chall_A['Total_Action_x'].value_counts(normalize=True).index.to_list()

fig, ax = plt.subplots(figsize=(8,6))

wedges, texts, autotexts = ax.pie(pie_data, labels=pie_label,
                                  startangle=90,
                                  autopct='%.0f%%',
                                  textprops={'color':'w', 'fontsize':16, 'weight':'bold'})

for i, wedge in enumerate(wedges):
  texts[i].set_color(wedge.get_facecolor())

plt.tight_layout()
```
